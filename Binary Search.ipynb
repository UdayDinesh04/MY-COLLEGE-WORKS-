```python
list=[] #created a empty list 
for i in range(0,101,2):#here we have directly taken range and difference(2) after that
    list.append(i) #it is expend list
    

def binarySearch(list,x,y,num):
    if y>=x:
        mid=x+(y-x)//2 # it is formula for mid element 
    
        if list[mid]==num:
            return mid
        elif list[mid] > num:
            return binarySearch(list, x, mid-1, num) #it is condition for when user input is less then mid element 
        else:
            return binarySearch(list, mid + 1, y, num)#it is condition for when user input is greater then mid element
    else:
        return -1 # Element is not present in the array
    
num=int(input("enter name :")) # user input 
#function call
result = binarySearch(list, 0, len(list)-1, num)
  
if result != -1:
    print("Element is found at index % d"%result)
else:
    print("Element is not found in list")

   
```
